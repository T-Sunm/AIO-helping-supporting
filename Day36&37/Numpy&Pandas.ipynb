# A. Numpy


### 1.1 Tạo mảng arr từ data và in ra kết quả 


```python
data = [30, 22, 47, 20, 46, 47, 37, 24, 40, 12, 43, 45, 44, 48, 24, 30, 15, 28, 21, 29, 17, 24, 41, 33, 23]
data = np.array(data)
start, end = 0, 5
print("Mảng arr: ",data)
print("5 phần tử đầu tiên của mảng", data[start:end])
print("5 phần tử cuối cùng của mảng", data[-1:-end - 1:-1])
```

### 1.2 Từ mảng 1 chiều arr, chuyển đổi thành mảng 2 chiều arr_2d có kích thước 5x5 và in mảng arr_2d ra màn hình

```python
arr_2d = data.reshape(5, -1)
print("Mảng 2 chiều: \n", arr_2d)
```

### 1.3 Lọc và in ra các phần tử có giá trị là số chẵn và số lẻ trong arr_2d


```python
arr_even = arr_2d[arr_2d % 2 == 0]
arr_odd = arr_2d[arr_2d % 2 != 0]

print("Các phần tử có giá trị là số chẵn: \n", arr_even)
print("Các phần tử có giá trị là số lẻ: \n", arr_odd)
```

### 1.4 Đếm số phần tử chẵn và số phần tử lẻ có trong mảng arr_2d


```python
print("Số phần tử có giá trị là số chẵn:", arr_even.shape[0])
print("Số phần tử có giá trị là số lẻ:", arr_odd.shape[0])
```

### 2. Đọc 2 tập tin heights.txt và position.txt vào 2 numpy array: np_positions và np_heights. 

```python
path = './numpy_data/'

path_height = path + 'heights.txt'
path_position = path + 'positions.txt'

# delimiter để cắt dấu , phân cách các phần tử trong txt
height_data = np.genfromtxt(path_height, delimiter=',', dtype=float)
position_data = np.genfromtxt(path_position, delimiter=',', dtype=str)
print(position_data)
```

### 2.1 Tính chiều cao trung bình của các GK ( Goal Keeper )

```python
cond_gk = position_data == "GK"
print(cond_gk)

GK_heights = height_data[cond_gk]
GK_heights.mean()
```

### 2.2 Tính chiều cao trung bình của các vị trí khác ( Không phải Goal keeper )

```python
all_position = np.unique(position_data)

all_position_without_gk = all_position[all_position != "GK"]
for pos in all_position_without_gk:
    pos_heights = height_data[position_data == pos].mean()
    print(pos_heights, end=', ')
```

### 2.3 Cho biết chiều cao lớn nhất và bé nhất của GK ( Goal Keeper )


```python
print("Chiều cao nhỏ nhất: ", GK_heights.min())
print("Chiều cao lớn nhất: ", GK_heights.max())
```

### 2.4 Tính chiều cao trung bình theo từng nhóm: GK (Goal Keeper ) , M ( Middle) , A (Attack ) , D ( Defense )

```python
all_position = zip(np.unique(position_data), ['Attack', 'Defense', 'Goal Keeper', 'Middle'])

for pos, name in all_position :
    print(f"{name} có chiều cao trung bình là: {height_data[position_data == pos].mean():.2f}",
          )
```

### 2. Vẽ biểu đồ thể hiện sự tương quan giữa Total Fat và Calories

```python
import seaborn as sns

sns.jointplot(data=menu_data, y='Calories', x='Total Fat', kind='scatter')
```

### 3 : mối quan hệ giữa 'Calories' và 'Total Fat' có vẻ là một tương quan thuận, nghĩa là khi giá trị 'Total Fat' tăng, giá trị 'Calories' cũng có xu hướng tăng theo.

### 4 : Hãy sắp xếp data theo thứ tự tăng dần theo Calories 

```python
menu_data.sort_values(inplace=True, by=["Calories"])

menu_data.head()
```

### 5 Hãy cho biết thực phẩm nào ( items ) có calories cao nhất

```python
print("Thực phẩm có calories cao nhất là:", menu_data.tail(1)["Item"].values[0])
```

### 6  Hãy cho biết nhóm thực phẩm nào ( Categories ) trong bảng thống kê này, chứa nhiều thực phẩm nhất ( items ) so với các loại còn lại

```python
product_category = menu_data.groupby("Category").count()["Item"]

# product_category.idxmax(): Trả về tên chỉ mục (tức là tên Category) của giá trị lớn nhất trong product_category.
print("Nhóm thực phẩm có nhiều items nhất là:", product_category.idxmax())
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  câu 1: đọc và hiển thị ảnh

img = cv2.imread ("1.jpg")
img1 = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img1)
plt.show()
```

```python
# Câu 2: Chuyển ảnh màu thành xám và hiển thị
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img2, cmap="gray")
plt.show()
```

```python
# img.astype(np.float32): Chuyển đổi kiểu dữ liệu của ảnh img từ uint8 (giá trị pixel từ 0 đến 255) sang float32 (giá trị pixel có thể lớn hơn 255 hoặc nhỏ hơn 0). Điều này được thực hiện để có thể thêm một giá trị số lớn hơn vào từng pixel mà không gặp lỗi tràn số.
newimg1 = img1.astype(np.float32) + 50
# đảm bảo giá trị nằm trong khoảng 0 -> 255
newimg1 = np.clip(newimg1, 0, 255)
# ảnh chỉ kh thể show ở dạng float --> chuyển về int
newimg1 = newimg1.astype(np.uint8)

newimg2 = img1.astype(np.float32) - 100
newimg2 = np.clip(newimg2, 0, 255)
newimg2 = newimg2.astype(np.uint8)


fig, axes = plt.subplots(2, 2, figsize=(10, 8))


# Plot each image in the grid
axes[0, 0].imshow(img1)
axes[0, 0].set_title('Origin Image')
axes[0, 0].axis('off')

axes[0, 1].imshow(newimg1)
axes[0, 1].set_title('Brightness Image')
axes[0, 1].axis('off')

axes[1, 0].imshow(newimg2)
axes[1, 0].set_title('Darkness Image')
axes[1, 0].axis('off')

axes[1, 1].axis('off')

plt.show()
```

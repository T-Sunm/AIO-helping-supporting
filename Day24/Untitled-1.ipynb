```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  cu 1: c v hin th nh

img = cv2.imread ("1.jpg")
img1 = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img1)
plt.show()
```

```python
# Cu 2: Chuyn nh mu thnh xm v hin th
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img2, cmap="gray")
plt.show()
```

```python
# img.astype(np.float32): Chuyn i kiu d liu ca nh img t uint8 (gi tr pixel t 0 n 255) sang float32 (gi tr pixel c th ln hn 255 hoc nh hn 0). iu ny c thc hin  c th thm mt gi tr s ln hn vo tng pixel m khng gp li trn s.
newimg1 = img1.astype(np.float32) + 50
# m bo gi tr nm trong khong 0 -> 255
newimg1 = np.clip(newimg1, 0, 255)
# nh ch kh th show  dng float --> chuyn v int
newimg1 = newimg1.astype(np.uint8)

newimg2 = img1.astype(np.float32) - 100
newimg2 = np.clip(newimg2, 0, 255)
newimg2 = newimg2.astype(np.uint8)


fig, axes = plt.subplots(2, 2, figsize=(10, 8))


# Plot each image in the grid
axes[0, 0].imshow(img1)
axes[0, 0].set_title('Origin Image')
axes[0, 0].axis('off')

axes[0, 1].imshow(newimg1)
axes[0, 1].set_title('Brightness Image')
axes[0, 1].axis('off')

axes[1, 0].imshow(newimg2)
axes[1, 0].set_title('Darkness Image')
axes[1, 0].axis('off')

axes[1, 1].axis('off')

plt.show()
```
